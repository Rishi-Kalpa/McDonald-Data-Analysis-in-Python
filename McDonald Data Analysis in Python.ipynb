# Q.04 Which category contributes to the maximum % of Cholesterol in a diet (% daily value)?

```python
McDonald.columns
```

```python
McDonald['Category'].nunique()
print('The number of Categories is -' , McDonald['Category'].nunique())
```

```python
McDonald_New = pd.pivot_table(McDonald, 'Cholesterol (% Daily Value)', index=['Category'])
Output1 = McDonald_New.sort_values(('Cholesterol (% Daily Value)'), ascending=False)
Output1
```

#### 'Breakfast' Category contributes to the maximum % of Cholesterol in a diet (% daily value)
